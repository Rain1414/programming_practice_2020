1) Дан список чисел. Определите, сколько в этом списке элементов, которые больше двух своих соседей, и выведите количество таких элементов. Крайние элементы списка никогда не учитываются, поскольку у них недостаточно соседей. (pythontutor.ru, занятие 7, «Больше своих соседей»)
1.png

```
a = input().split()
count = 0
for i in range(len(a)):
    a[i] = int(a[i])
for i in range(1, len(a) - 1):
    if (a[i] > a[i - 1]) and (a[i] > a[i + 1]):
        count += 1
print(count)
```

2) В списке все элементы различны. Поменяйте местами минимальный и максимальный элемент этого списка. (pythontutor.ru, занятие 7, «Переставить min и max»)
2.png

```
a = input().split()
for i in range(len(a)):
    a[i] = int(a[i])
i_min = a.index(min(a))
i_max = a.index(max(a))
a[i_min], a[i_max] = a[i_max], a[i_min]
for i in a:
    print(i)
```

3) Дан список чисел. Посчитайте, сколько в нем пар элементов, равных друг другу. Считается, что любые два элемента, равные друг другу образуют одну пару, которую необходимо посчитать. (pythontutor.ru, занятие 7, «Количество совпадающих пар»)
3.png

```
a = input().split()
for i in range(len(a)):
    a[i] = int(a[i])
sum = 0
for i in range(len(a)):
    if a[i] in a[:i]:
        continue
    if a.count(a[i]) > 1:
        sum += a.count(a[i]) * (a.count(a[i]) - 1) / 2
print(int(sum)) 
```

8) Вам дан словарь, состоящий из пар слов. Каждое слово является синонимом к парному ему слову. Все слова в словаре различны. Для слова из словаря, записанного в последней строке, определите его синоним. (pythontutor.ru, занятие 11, «Словарь синонимов»)
8.png

```
dict_ = dict()
for i in range(int(input())):
    items = input().split()
    dict_[items[0]] = items[1]
a = input()
if a in dict_.keys():
    print(dict_[a])
else:
    for i in dict_:
        if dict_[i] == a:
            print(i)
```

9) Дан текст: в первой строке задано число строк, далее идут сами строки. Выведите слово, которое в этом тексте встречается чаще всего. Если таких слов несколько, выведите то, которое меньше в лексикографическом порядке. (pythontutor.ru, занятие 11, «Самое частое слово»)
9.png

```
dict_ = dict()
max_ = 0
for i in range(int(input())):
    for word in input().split():
        if dict_.get(word) != None:
            dict_[word] += 1
        else:
            dict_.update({word: 1})
        if dict_[word] > max_:
                max_ = dict_[word]
list_max = list()
for key_ in dict_:
    if dict_[key_] == max_:
        list_max.append(key_)
list_max.sort()
print(list_max[0])
```

9) Дана база данных о продажах некоторого интернет-магазина. Каждая строка входного файла представляет собой запись вида Покупатель товар количество, где Покупатель — имя покупателя (строка без пробелов), товар — название товара (строка без пробелов), количество — количество приобретенных единиц товара. Создайте список всех покупателей, а для каждого покупателя подсчитайте количество приобретенных им единиц каждого вида товаров. Список покупателей, а также список товаров для каждого покупателя нужно выводить в лексикографическом порядке. (pythontutor.ru, занятие 11, «Продажи»)
10.png

```
dict_ = dict()
dict_final = dict()
while 0 == 0:
    try:
        a = input().split()
        if dict_.get(a[0]) != None:
            if a[1] in dict_[a[0]].keys():
                dict_[a[0]][a[1]] += int(a[2])
            else:
                dict_[a[0]].update({a[1]: a[2]})
        else:
            d = {a[1]: int(a[2])}
            dict_.update({a[0]: d})
    except:
        break
names = sorted(dict_.keys())
for name in names:
    dict_final.update({name: dict()})
    products = sorted(dict_[name])
    for product in products:
        dict_final[name].update({product: dict_[name][product]})
for name in dict_final:
    print(name+":")
    for product in dict_final[name]:
        print(product, dict_final[name][product])
```
