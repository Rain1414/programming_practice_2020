1) Дано действительное положительное число a и целоe число n. Вычислите a в степени n. Решение оформите в виде функции power (a, n). Стандартной функцией возведения в степень пользоваться нельзя. (pythontutor.ru, 8 урок, отрицательная степень)
1.png

```
def power_(a: int, n: int):
  """
  Возводит число a в в степень n.
  :param a: действительное положительное число
  :param n: целоe число
  :return: a^n
  """
  # Задаём начальное число, от которого начнём считать результат. Оно равно а
  pow_ = a
  # Если степень равна нулю, то просто выводим 1
  if n == 0:
      return 1
  # Для положительных степеней умножаем наше итоговое число pow_ на а столько раз,
  # чему равно n 
  elif n > 0:
      while n > 1:
          pow_ *= a
          n -= 1
      return pow_
  # Для отрицательных степеней умножаем наше итоговое число pow_ на а столько раз,
  # чему равно n, и берём обратное от этого итогового pow_ 
  else:
      while n < -1:
            pow_ *= a
            n += 1
      return 1 / pow_
print(power_(float(input()), int(input())))
```

2) Напишите функцию вычисляющие наибольшие общие делители (НОД) для множества пар чисел.

```
# Задаём числа
a, b = int(input()), int(input())

# (пользовался алгоритмом с википедии)
# Проводим цикл, пока одно из них не занулится
while a != 0 and b != 0:
    # сравниваем числа. Если a больше, то надо найти  остаток от деления a на b 
    # и записываем в a
    if a > b:
        a = a % b
    # Иначе делаем наоборот. Делим b на а и присваеваем результат к b
    else:
        b = b % a
# В итоге одно число будет равно нулю, а другое - наш НОД
# Возвращем максимум
print(max(a, b))
```

3) Напишите функцию, которая в зависимости от выбора пользователя вычисляет площадь круга, прямоугольника или треугольника. Для вычисления площади каждой фигуры должна быть написана отдельная функция.

```
from numpy import pi, sin
def circle_square(radius: float):
    """
    Вычисляет площадь круга
    :param radius: радиус круга
    :return: площадь фигуры
    """
    return pi * radius * radius
def rectangle_square(a: float, b: float):
    """
    Вычисляет площадь прямоугольника
    :param a: одна сторона фигуры
    :param b: вторая сторона фигуры
    :return: площадь фигуры
    """
    return a * b
def triangle_square(a: float, b: float, angle: float):
    """
    Вычисляет площадь прямоугольника
    :param a: одна сторона фигуры
    :param b: вторая сторона фигуры
    :return: площадь фигуры
    """
    return 0.5 * a * b * sin(angle)
print("Choose figure: Circle - 1, rectangle - 2, triangle - 3")
# Выполняем функцию в соответствии с указанным номером
i = int(input())
if i == 1:
    print("Enter the radius")
    print(circle_square(float(input())))
elif i == 2:
    print("Enter the sides")
    print(rectangle_square(float(input()), float(input())))
elif i == 3:
    print("Enter the two sides and then the angle between them in radians")
    print(triangle_square(float(input()), float(input()), float(input())))
else:
    print("Please, enter 1 or 2 or 3")
```

Вычисление значения функции оформить в виде программной функции, которая принимает значение x, а возвращает полученное значение функции (y). Нарисуйте с помощью matplotlib график этой функции.



```
import matplotlib.pyplot as plt

def func(x: float):
    if -10 <= x < -5:
        return 2 * abs(x) - 1
    elif -5 <= x <= 5:
        return x * x
    elif 5 < x <= 10:
        return 2 * x
    else:
        return
a = []
for x in range(-10, 11, 1):
    a.append(func(x))
plt.plot(range(-10, 11, 1), a)
```
